```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import mplcursors
# %matplotlib notebook
```

```python
def make_df(url):
    df = pd.read_csv(url, encoding='latin-1')
    tri_states = ['NY', 'NJ', 'CT']
    df = df.loc[df['State'].isin(tri_states)].reset_index()
    cols_wanted = ['RegionName', 'City', 'State', 'CountyName' , 
                   '2014-01','2015-01','2016-01','2017-01', '2018-01', '2019-01']
    df = df[cols_wanted]
    df.rename(columns = {'2014-01': 2014,'2015-01': 2015,'2016-01': 2016,
        '2017-01': 2017, '2018-01': 2018, '2019-01': 2019}, inplace=True)
    return df
```

```python
ratio_df = make_df('CSV/Zip_PriceToRentRatio_AllHomes.csv')
ratio_df
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,8))
sns.distplot(ratio_df[2019], color = 'green')
plt.title('Price to Rent Ratio For 2019\n Distrubution Plot', size = 20)
```

```python
def plot_state_ratio(state):
    for_plot = {}
    count = []
    groups = ratio_df[ratio_df['State'] == state].groupby(['CountyName'])
    for name, group in groups:
        for_key = name.split(' ')
        key = ' '.join(for_key[0:-1])
        for_plot[key] = group[2019].median()
        count.append(len(group))
    plt.figure(figsize=(20,10))
    plt.bar(for_plot.keys(), for_plot.values(), color = [
        'PeachPuff', 'LightSteelBlue', 'teal'])
    plt.title(f'Median Price to Rent Ratio by County for {state}', size=15)
    plt.savefig('test2.png')
    # plt.text(range(len(for_plot)))
    # for i, v in enumerate(count, ):
    #     plt.text(for_plot[i] ,v , str(v), color='blue', fontweight='bold')
    for i, v in enumerate(for_plot.values()):
        plt.text( i-0.15 ,v + 0.25 , str(count[i]), color='SlateBlue',
                 size=15, fontweight='bold')
    plt.show()
```
